```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix

# Mapeamento de números para categorias
numero_para_categoria = {
    0: 'BACKGROUND',
    1: 'BULK-CARRIER',
    2: 'CONTAINER-SHIP',
    3: 'GENERAL-CARGO',
    4: 'OIL-PRODUCTS-TANKER',
    5: 'PASSENGERS-SHIP',
    6: 'TANKER',
    7: 'TRAWLER',
    8: 'TUG',
    9: 'VEHICLES-CARRIER',
    10: 'YACHT'
}

def plot_confusion_matrix(confusion_matrix_data, class_names, title='Confusion Matrix', cmap=plt.cm.Blues):
    plt.figure(figsize=(12, 10))
    sns.set(font_scale=1.2)

    # Calculando as porcentagens
    total_samples_per_class = np.sum(confusion_matrix_data, axis=1)
    # porcentagem de 0 a 1
    confusion_matrix_data_percentage = confusion_matrix_data / total_samples_per_class[:, np.newaxis]

    sns.heatmap(confusion_matrix_data_percentage, annot=True, fmt='.2f', cmap=cmap)
    plt.title(title)
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')

    # Definindo os rótulos para os eixos x e y
    plt.xticks(ticks=np.arange(len(class_names)), labels=class_names, rotation=90)
    plt.yticks(ticks=np.arange(len(class_names)), labels=class_names, rotation=0)

    plt.show()

def calculate_metrics(json_data):
    for dtype, data in json_data['validation'].items():
        print(f"Metrics for data type: {dtype}")
        confusion_matrix_data = np.array(data['confusion_matrix'])
        class_names = [numero_para_categoria[i] for i in range(len(confusion_matrix_data))]

        # 1. Matriz de Confusão
        plot_confusion_matrix(confusion_matrix_data, class_names, title=f'Confusion Matrix - {dtype}')

        # 2. Precisão
        precision = data['non_background']['precision']
        print(f"\nPrecision: {precision}")

        # 3. Revocação
        recall = data['non_background']['recall']
        print(f"\nRecall: {recall}")

        # 4. F1-score
        f1_score = data['non_background']['f1']
        print(f"\nF1-score: {f1_score}")

        # 5. Classification Report
        print("\nClassification Report:")
        classification_report_dict = data['classification_report']
        print(classification_report_dict)

# Chamada da função
calculate_metrics(json_data)
```
